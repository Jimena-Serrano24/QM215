```python
def generarDias(nEstudiantes):
  dias=np.random.randint(1, 366,size=nEstudiantes)
  op=False
  for i in range(0,20):
    n1=(dias[i])
    v=(dias[i+1:])
    for j in range(0,len(v)):
      if n1==v[j]:
        op=True
  return op
```

```python
n=0
for i in range(1000):
  if generarDias(20):
    n+=1
print(n/1000*100)
```

```python
A=1.0
P=0.0
K=0.1
t=0
dt=0.01
tt=[]
AA=[]
PP=[]
while t<30:
    tt.append(t)
    AA.append(A)
    PP.append(P)
    dA=-K*A*dt
    dP=K*A*dt
    A=A+dA
    P=P+dP
    t+=dt
tt=np.array(tt)
AA=np.array(AA)
PP=np.array(PP)

x=np.arange(0,50,5)
y=np.exp(-K*x)
y2=1-y

plt.plot(tt,AA)
plt.plot(tt,PP)

plt.scatter(x,y,color="Fuchsia",s=8)
plt.scatter(x,y2,color="Fuchsia",s=8)

plt.xlabel("t(seg)")
plt.ylabel("[A]")
plt.title("Reacción Primer Orden A-->P")


plt.grid(True)
plt.show()
```

```python
A=1.0
P=0.0
K=0.1
t=0
dt=0.01
tt=[]
AA=[]
PP=[]
while t<30:
    tt.append(t)
    AA.append(A)
    PP.append(P)
    dA=-K*A*A*dt
    dP=K*A*A*dt
    A=A+dA
    P=P+dP
    t+=dt
tt=np.array(tt)
AA=np.array(AA)
PP=np.array(PP)

x=np.arange(0,50,2)
A0=1.0
invAt=K*x+1/A0
y=1/invAt

x=np.arange(0,50,5)
y=np.exp(-K*x)
y2=1-y

plt.plot(tt,AA)
plt.plot(tt,PP)

plt.scatter(x,y,color="Fuchsia",s=8)

plt.xlabel("t(seg)")
plt.ylabel("[A]")
plt.title("Reacción Segundo Orden A-->P")


plt.grid(True)
plt.show()
```

```python
A=1.0
P=0.0
X=0.0
k1=0.1
k2=0.2
t=0
dt=0.1
tt=[]
AA=[]
PP=[]
XX=[]
rojo_puro = '#FF0000'
azul_puro = '#0000FF'

while t<50:
  tt.append(t)
  AA.append(A)
  PP.append(P)
  XX.append(X)
  dA=-k1*A*dt
  dX=k1*A*dt-k2*X*dt
  dP=k2*X*dt
  P=P+dP
  A=A+dA
  t+=dt
  X=X+dX

tt=np.array(tt)
AA=np.array(AA)
PP=np.array(PP)
XX=np.array(XX)

plt.plot(tt,AA)
plt.plot(tt,XX)
plt.plot(tt,PP)
plt.xlabel("t(seg)")
plt.ylabel("[A]")
plt.grid(True)
plt.title("Reacción consecutiva orden A-->X    x--> p ")
plt.show()
```
